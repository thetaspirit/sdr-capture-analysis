```python
# includes core parts of numpy, matplotlib
import matplotlib.pyplot as plt
import numpy as np
# include scipy's signal processing functions
import scipy.signal as signal
```

### Defining Functions and Constants

```python
TARGET_DIR = "E:\\"

FFT_SIZE = 2048
SAMPLE_RATE = 61.44e6
```

```python
def loadData(filepath):
    dat = np.fromfile(filepath, dtype="float32")
    dat = dat[0::2] + 1j*dat[1::2]
    return dat

def plotSpecgram(dat):
    plt.specgram(dat, NFFT=FFT_SIZE, Fs=SAMPLE_RATE)
    plt.gcf().set_figwidth(20)
    plt.title("Spectogram of 'signal' loaded from file")
    plt.xlabel("Time")
    plt.ylabel("Frequency")
    plt.show()

def plotPSD(dat):
    plt.psd(dat, NFFT=FFT_SIZE, Fs=SAMPLE_RATE)
    plt.gcf().set_figwidth(20)
    plt.title("PSD of 'signal' loaded from file")
    plt.show()

def plotConstellation(dat):
    plt.scatter(np.real(dat[0:200000]), np.imag(dat[0:200000]), 3)
    plt.gcf().set_figwidth(20)
    plt.gcf().set_figheight(20)
    plt.title("Constellation of the 'signal' loaded from file")
    plt.show()
```

## Normal Communication
between camera and monitor

```python
dat = loadData(f"{TARGET_DIR}normal")
plotSpecgram(dat)
plotPSD(dat)
plotConstellation(dat)

normalCaputreSize = len(dat)
```

### Both devices off

```python
dat = loadData(f"{TARGET_DIR}nothing")
plotSpecgram(dat)
plotPSD(dat)
plotConstellation(dat)
```

```python
# limiting capture size to the size of "normal"'s capture
newdat = dat[normalCaputreSize:]
plotSpecgram(newdat)
plotPSD(newdat)
plotConstellation(newdat)
```
